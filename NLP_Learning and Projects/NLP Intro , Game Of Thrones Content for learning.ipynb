```python
import kagglehub
from pandas.io.xml import preprocess_data

# Download latest version
path = kagglehub.dataset_download("khulasasndh/game-of-thrones-books")

print("Path to dataset files:", path)
```

```python
import os
from nltk.tokenize import RegexpTokenizer

file_path = os.path.join(path, "001ssb.txt")

with open(file_path, encoding="utf-8") as f:
    sub_data = f.read(10000)

tokenizer = RegexpTokenizer(r'\w+')
tokens = tokenizer.tokenize(sub_data)

print(tokens)
```

```python
import os
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.stem import WordNetLemmatizer
from gensim.models import Word2Vec
```

```python
import spacy
nlp = spacy.load('en_core_web_sm')
```

```python
all_lemmas = []
for sentence in tokens:
    doc = nlp(sentence)
    lemmas = [token.lemma_ for token in doc]
    all_lemmas.append(lemmas)
    print(lemmas)
```

```python
def chunk_list(lst, n):
    """Yield successive n-sized chunks from lst."""
    for i in range(0, len(lst), n):
        yield lst[i:i + n]

# For example, group tokens into "sentences" of 20 words
sentences = list(chunk_list(tokens, 20))

from gensim.models import Word2Vec

model = Word2Vec(
    sentences=sentences,
    vector_size=100,
    window=5,
    min_count=1,
    workers=4,
    sg=1
)

# Test: Get similar words or vector for any word
print(model.wv.most_similar('man'))
print(model.wv['man'])
```
